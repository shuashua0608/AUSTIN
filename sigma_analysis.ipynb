```python
epsilon = 1e-9  # Small value to avoid log(0)
sigma_values = np.concatenate(sigma_values) if isinstance(sigma_values, list) else np.array(sigma_values)
mri_labels = np.array(mri_labels, dtype=np.int64) 
tri_labels = np.array(tri_labels, dtype=np.int64)

cross_entropy = -np.mean(
    mri_labels * np.log(sigma_values + epsilon) + (1 - mri_labels) * np.log(1 - sigma_values + epsilon)
)

# Step 5 (Optional): Compute Weighted Loss Based on Consistency
weighted_cross_entropy = -np.sum(
    consistency * (mri_labels * np.log(sigma_values + epsilon) + (1 - mri_labels) * np.log(1 - sigma_values + epsilon))
) / np.sum(consistency)  # Normalize by total consistent cases

# Output results
print(f"Cross-Entropy Loss: {cross_entropy:.4f}")
print(f"Weighted Cross-Entropy Loss (Optional): {weighted_cross_entropy:.4f}")
```

```python

# Step 1: Compute Consistency
consistency = (mri_labels == tri_labels).astype(int)  # 1 if consistent (00, 11), 0 if inconsistent (01, 10)

# Step 2: Separate Sigma Values Based on Consistency
sigma_consistent = sigma_values[consistency == 1]  # Consistent cases (00, 11)
sigma_inconsistent = sigma_values[consistency == 0]  # Inconsistent cases (01, 10)

# Step 3: Compute Statistics
print(f"Consistent Cases (00, 11) - Mean Sigma: {np.mean(sigma_consistent):.4f}, Std: {np.std(sigma_consistent):.4f}")
print(f"Inconsistent Cases (01, 10) - Mean Sigma: {np.mean(sigma_inconsistent):.4f}, Std: {np.std(sigma_inconsistent):.4f}")

# Step 4: Boxplot to Compare Distributions
plt.figure(figsize=(8, 5))
sns.boxplot(x=consistency, y=sigma_values, palette=["blue", "red"])
plt.xticks([0, 1], ["Inconsistent (01, 10)", "Consistent (00, 11)"])
plt.xlabel("MRI & Triage Consistency")
plt.ylabel("Uncertainty Score (Sigma)")
plt.title("Uncertainty Scores for Consistent vs. Inconsistent Cases")
plt.show()

# Step 5: Further Breakdown by Label Cases (00, 11, 01, 10)
label_groups = {"00": [], "11": [], "01": [], "10": []}
for i in range(len(mri_labels)):
    label_key = f"{mri_labels[i]}{tri_labels[i]}"
    label_groups[label_key].append(sigma_values[i])

# Convert to NumPy arrays
for key in label_groups:
    label_groups[key] = np.array(label_groups[key])

# Print stats for each category
for key in label_groups:
    print(f"Case {key} - Mean Sigma: {np.mean(label_groups[key]):.4f}, Std: {np.std(label_groups[key]):.4f}, Count: {len(label_groups[key])}")

# Step 6: Boxplot for Each Category
plt.figure(figsize=(8, 5))
sns.boxplot(data=[label_groups["00"], label_groups["11"], label_groups["01"], label_groups["10"]],
            palette=["green", "blue", "orange", "red"])
plt.xticks([0, 1, 2, 3], ["00", "11", "01", "10"])
plt.xlabel("MRI Label - Triage Label")
plt.ylabel("Uncertainty Score (Sigma)")
plt.title("Uncertainty Scores for Each MRI-Triage Case")
plt.show()

# Step 7: Statistical Test - Compare Consistent vs. Inconsistent Cases
t_stat, p_value = ttest_ind(sigma_consistent, sigma_inconsistent, equal_var=False)  # Welchs t-test
print(f"T-Test between Consistent and Inconsistent Cases: t={t_stat:.4f}, p={p_value:.4e}")
```
